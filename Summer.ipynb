# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
f1=pd.read_csv('summer.csv')
df=pd.DataFrame(f1)
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
d1=df[df['Medal']=='Gold']
lst=[]
for i in d1['Sport'].unique():
    lst.append([i,len(d1[d1['Sport']==i])])
d1=pd.DataFrame(lst,columns=['Sport','Gold_Medals']).sort_values(by='Gold_Medals',ascending=False).head()

print(d1['Sport'][0],",this Sport is having most number of Gold Medals so far.")
d1.plot(x='Sport',y='Gold_Medals',kind='bar')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
lst=[]
for i in df['Sport'].unique():
    lst.append([i,len(df[df['Sport']==i])])
    
d1=pd.DataFrame(lst,columns=['Sport','Medals']).sort_values(by='Medals',ascending=False).head()
print(d1['Sport'][0],",this Sport is having most number of Medals so far.")

d1.plot(x='Sport',y='Medals',kind='bar')
```

### 4. Which player has won most number of medals? (Top 5)

```python
lst1=[]
for i in df['Athlete'].unique():
    lst1.append([i,len(df[df['Athlete']==i])])
    
d1=pd.DataFrame(lst1,columns=['Athlete','Athlete Most Medals Winner']).sort_values(by='Athlete Most Medals Winner',ascending=False).head()

print('"',d1['Athlete'][18640],'"',",this Player has won most number of Medals.")
d1.plot(x='Athlete',y='Athlete Most Medals Winner',kind='bar')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
d3=df[df['Medal']=='Gold']
lst2=[]
for i in d3['Athlete'].unique():
    lst2.append([i,len(d3[d3['Athlete']==i])])
d3=pd.DataFrame(lst2,columns=['Athlete','Athlete Most Gold_Medals Winner']).sort_values(by='Athlete Most Gold_Medals Winner',ascending=False).head()

print('"',d3['Athlete'][6851],'"',",this Player has won most number of Gold Medals.")
d3.plot(x='Athlete',y='Athlete Most Gold_Medals Winner',kind='bar')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
d1=df[df['Medal']=='Gold']
d1=d1[d1['Country']=='IND']
year=d1['Year'].unique().min()
print("India won First Gold medal in year : ",year) 
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
lst1=[]
for i in df['Sport'].unique():
    lst1.append([i,len(df[df['Sport']==i])])
d1=pd.DataFrame(lst1,columns=['Sport','Top 5 Popular Sports']).sort_values(by='Top 5 Popular Sports',ascending=False).head()
print(d1['Sport'][0],",this event is Most popolar")

d1.plot(x='Sport',y='Top 5 Popular Sports',kind='bar')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
d1=df[df['Medal']=='Gold']
d1=d1[d1['Gender']=='Women']
lst=[]
for sport in d1['Sport'].unique():
    lst.append([sport,len(d1[d1['Sport']==sport])])
d1=pd.DataFrame(lst,columns=['Sport','Female_Gold_Medalist']).sort_values(by='Female_Gold_Medalist',ascending=False).head()
print(d1['Sport'][4],',this Sport is having Most number of female Gold Medalists.')

d1.plot(x='Sport',y='Female_Gold_Medalist',kind='bar')
```
